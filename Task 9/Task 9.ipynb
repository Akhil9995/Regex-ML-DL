```python
test_id = test.pop('id')
```

```python
trainy=train.pop('species')
```

```python
train.drop(['id'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder().fit(trainy)
Y = encoder.transform(trainy)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(train)
test=sc.fit_transform(test)
```

```python
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state =1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 40,criterion = 'entropy',max_depth = 20,random_state = 5)
rf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

print('training accuracy = ',accuracy_score(y_train,rf.predict(x_train)))
print('testing accuracy = ',accuracy_score(y_test,rf.predict(x_test)))
```

```python
y_pred=rf.predict(test)
```

```python
results=pd.DataFrame()
```

```python
ls=list(encoder.classes_)
```

```python
rslts=[ls[i] for i in y_pred]
```

```python
results['id']=test_id
results['species']=rslts
```

```python
results.set_index('id')
```

```python
results.to_csv('submission_leaf_classification.csv')
```
